```r
library("PerformanceAnalytics")
library(tidyverse)
library(readxl)
```

```r
returns = as.data.frame(read_excel("excess.xlsx"))
```

```r
returns
```

```r
CAPM = xts(returns[,2:length(returns)], order.by=as.Date(returns[,1]))
```

```r
CAPM.beta(Ra=CAPM$`Excess GS US Equity Return`, Rb=CAPM$MR_RF)
CAPM.alpha(Ra=CAPM$`Excess GS US Equity Return`, Rb=CAPM$MR_RF)
summary(lm(returns$`Excess GS US Equity Return`~returns$MR_RF))$r.squared
```

```r
summary(lm(returns$`Excess GS US Equity Return`~returns$MR_RF))
```

```r
res = data.frame(r2=0.0, alpha=0.0, beta=0.0, p=0.0)
for(i in 1:(ncol(CAPM) - 4)) {
    res[i, 3] = CAPM.beta(Ra=CAPM[, i], Rb=CAPM$MR_RF)
    res[i, 2] = CAPM.alpha(Ra=CAPM[, i], Rb=CAPM$MR_RF)
    res[i, 1] = summary(lm(CAPM[, i]~CAPM$MR_RF))$r.squared
    res[i, 4] = summary(lm(CAPM[, i]~CAPM$MR_RF))$coefficients[2,4]
}
```

```r
res
```

```r
# Fama French
```

```r
fama = data.frame(r2=0.0, alpha=0.0, beta_m=0.0,beta_smb=0.0, beta_hml=0.0, p_m=0.0, p_smb=0.0, p_hml=0.0)
for(i in 1:(ncol(CAPM) - 4)) {
    coeff = summary(lm(CAPM[, i]~CAPM$MR_RF + CAPM$SMB + CAPM$HML))
    fama[i, 1] = coeff$r.squared
    fama[i, 2] = coeff$coefficients[1, 1]
    fama[i, 3] = coeff$coefficients[2, 1]
    fama[i, 4] = coeff$coefficients[3, 1]
    fama[i, 5] = coeff$coefficients[4, 1]
    fama[i, 6] = coeff$coefficients[2, 4]
    fama[i, 7] = coeff$coefficients[3, 4]
    fama[i, 8] = coeff$coefficients[4, 4]
}
```

```r
summary(lm(CAPM[, 2]~CAPM$MR_RF + CAPM$SMB + CAPM$HML))
```

```r
fama
```

```r
carhart = data.frame(r2=0.0, alpha=0.0, beta_m=0.0,beta_smb=0.0, beta_hml=0.0, beta_mom=0.0, p_m=0.0, p_smb=0.0, p_hml=0.0, p_mom=0.0)
for(i in 1:(ncol(CAPM) - 4)) {
    coeff = summary(lm(CAPM[, i]~CAPM$MR_RF + CAPM$SMB + CAPM$HML+ CAPM$MOM))
    carhart[i, 1] = coeff$r.squared
    carhart[i, 2] = coeff$coefficients[1, 1]
    carhart[i, 3] = coeff$coefficients[2, 1]
    carhart[i, 4] = coeff$coefficients[3, 1]
    carhart[i, 5] = coeff$coefficients[4, 1]
    carhart[i, 6] = coeff$coefficients[5, 1]
    carhart[i, 7] = coeff$coefficients[2, 4]
    carhart[i, 8] = coeff$coefficients[3, 4]
    carhart[i, 9] = coeff$coefficients[4, 4]
    carhart[i, 10] = coeff$coefficients[5, 4]
}
```
